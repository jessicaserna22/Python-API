```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import urllib

from tqdm import tqdm
# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
    
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500) #size determins how many points we want to generate for this range, ask for 1500 because a lot of the points the majority of the world is ocean
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)# want to create tuples using zip, zipping together the list in the ()

# Identify nearest city for each lat, lng combination
#use citipy to find nearest city to lat, lng pairs
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list by using appen
    #does not include duplicates 
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count, to make sure it's greater than 500
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
#API calls for cities

api_key = api_keys.api_key 
print(api_key)

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/forecast?id=524901&APPID=" + api_key 
url
```

```python
#create list to hold data for cities
#adding dictionaries to these list


city_data = []

#loop through cities to create a query url and get a response
MissingDataCount = 0




for city in cities:
         
    query_url = url + "&q=" + city #the parameter for query is the middle part
#this will loop throught cities
# to debug can use 
#print(f"processing {city})
    
    #time.sleep(1)
      
    try:#officially makes request to API, look at how the API displays data in the webpage
        city_weather = requests.get(query_url).json()
        
        #parse json
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]
        
        #append city information
        city_data.append({"City": city,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
                      
                                         
       
    except:
        print("City not found. Skipping")
        MissingDataCount = MissingDataCount + 1
        pass
        
print("Incomplete Data for", int(MissingDataCount), "cities.")

City_Data = pd.DataFrame(city_data)
City_Data.head()
```
